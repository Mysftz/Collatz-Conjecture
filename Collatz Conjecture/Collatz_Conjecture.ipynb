```python
import numpy as np
import matplotlib.pyplot as plt

n = np.random.randint(10)
count = 0
x = [0]
y = []
y.append(n)

def collatz(z):
	# Unification of 3x+1 for odd and x/2 for even
	return 0.25*(2 + 7*z + (-1)*(2 + 5*z)*((-1) ** z))

while y[-1] != 1.0:
	count += 1
	x.append(count)
	y.append(collatz(y[-1]))
	#print(y[-1], end=", ")

print("Intial values: ",n)
print("Sequence values: ",y)

plt.plot(x, y)
plt.xlabel("Number of Iterations")
plt.ylabel("Sequence Values (n)")
plt.title(f"Graphical representation of Collatz Conjecture for n={n}")
plt.show()
```
